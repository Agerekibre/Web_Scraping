# Scraping `blocket.se` website

# Comparing Diesel and Electric Cars

## Diesel Cars

    Scraping HTML data
    Cleaning the data
    Checking the length of each required parameter
    Putting the data into pandas DataFrame

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

diesel_price = []
diesel_model = []
diesel_year  = []
diesel_area  = []
diesel_saler = []

Dl = pd.DataFrame()

for k in range (1,150,1):
    url_diesel = f"https://www.blocket.se/annonser/stockholm/fordon/bilar?ccca=1&ccce=1&ccco=1&cchb=1&ccsc=1&ccsd=1&ccsu=1&cg=1020&fu=2&page={k}&r=11"
    
    get = requests.get(url_diesel)
    time.sleep(2)
    html = get.content
    soup = BeautifulSoup(html)
    

    
    container = soup.select('div.styled__Wrapper-sc-1kpvi4z-0')

    for d in container:

        m = [d.text for d in d.select('h2.TextSubHeading__TextSubHeadingWrapper-sc-1ilszdp-0')]
        diesel_model.append(''.join(m))

        y = [d.text for d in d.select('ul.ParametersList__List-sc-18ndpo4-1 :nth-child(1)')]
        diesel_year.append(''.join(y))
        
        p = [d.text for d in d.select('div.TextSubHeading__TextSubHeadingWrapper-sc-1ilszdp-0')]
        diesel_price.append(''.join(p))  

        a = [d.text for d in d.select('p.styled__TopInfoWrapper-sc-1kpvi4z-24 :nth-child(2)')]
        diesel_area.append(''.join(a))

        s = [d.text for d in d.select('span.AdvertiserTypeSymbol__Badge-sc-54a824-0')]
        if len(s) == 0:
            diesel_saler.append('Privat')
        else:
            diesel_saler.append(''.join(s))
            
            
print(len(diesel_model))
print(len(diesel_year))
print(len(diesel_price))
print(len(diesel_area))
print(len(diesel_saler))


Dl['Diesel_Model'] = diesel_model
Dl['Diesel_Year']  = diesel_year
Dl['Diesel_Price'] = diesel_price
Dl['Diesel_Area']  = diesel_area
Dl['Diesel_Saler'] = diesel_saler
Dl
```

## Electric Cars

    Scraping HTML data
    Cleaning the data
    Checking the length of each required parameter
    Putting the data into pandas dataframe

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

electric_model = []
electric_year  = []
electric_price = []
electric_area  = []
electric_saler = []

El = pd.DataFrame()

for k in range (1,20,1):
    url_electric = f"https://www.blocket.se/annonser/stockholm/fordon/bilar?ccca=1&ccce=1&ccco=1&cchb=1&ccsc=1&ccsd=1&ccsu=1&cg=1020&fu=4&page={k}&r=11"
    get = requests.get(url_electric)
    time.sleep(2)
    html = get.content
    soup = BeautifulSoup(html)
    
    container = soup.select('div.styled__Wrapper-sc-1kpvi4z-0')

    for e in container:
        
        m = [e.text for e in e.select('div.styled__SubjectWrapper-sc-1kpvi4z-14 ')]
        electric_model.append(''.join(m))

        y = [e.text for e in e.select('ul.ParametersList__List-sc-18ndpo4-1 :nth-child(1)')]
        electric_year.append(''.join(y))

        p = [e.text for e in e.select('div.TextSubHeading__TextSubHeadingWrapper-sc-1ilszdp-0 ')]
        electric_price.append(''.join(p))

        a = [e.text for e in e.select('p.styled__TopInfoWrapper-sc-1kpvi4z-24 :nth-child(2)')]
        electric_area.append(''.join(a))

        s = [e.text for e in e.select('span.AdvertiserTypeSymbol__Badge-sc-54a824-0')]
        if len(s) == 0:
            electric_saler.append('Privat')
        else:
            electric_saler.append(''.join(s))
            
print(len(electric_model))
print(len(electric_year))
print(len(electric_price))
print(len(electric_area))
print(len(electric_saler))

El['Electric_Model'] = electric_model
El['Electric_Year']  = electric_year
El['Electric_Price'] = electric_price
El['Electric_Area']  = electric_area
El['Electric_Saler'] = electric_saler
El
```
